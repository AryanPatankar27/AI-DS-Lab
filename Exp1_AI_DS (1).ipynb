```python
df.reset_index(level=0,inplace=True)
df.head()
```

```python
cols=['Name','Ticket','Cabin']
df=df.drop(cols,axis=1)
df
```

```python
dummies=[]
cols=['Pclass','Sex','Embarked']
for col in cols:
  dummies.append(pd.get_dummies(df[col]))

heart_dummies=pd.concat(dummies,axis=1)
df=pd.concat((df,heart_dummies),axis=1)

df.info()
```

```python
#Filling the missing data in numerical columns with its median
df['Age'].fillna(df['Age'].median())
df.shape
df.isnull().sum()
```

```python
#attribute embarked does not affect the data so remove it

df
```

```python
#Box Plot
plt.figure(figsize=(16,8)) #(width,height)
sns.boxplot(df['PassengerId'])
```

```python
#Position of the outlier's index
print(np.where(df['Fare']>200))
```

```python
#Finding Quartiles and IQR
Q1=df.select_dtypes(include=np.number).quantile(0.25)
Q3=df.select_dtypes(include=np.number).quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
IQR["Fare"],Q1["Fare"],Q3["Fare"]
```

```python
#calculate the outlier cutoff based on the engine HP
cut_off=IQR["Fare"]*1.5
lower,upper=Q1["Fare"]-cut_off,Q3["Fare"]+cut_off
cut_off,lower,upper
```

```python
#getting the outliers
outliers=df[(df["Fare"]<lower)|(df["Fare"]>upper)]
outliers.head(10)
```
